```python
from skimage import io, transform
import matplotlib.pyplot as plt
import numpy as np
from skimage.color import rgb2gray
from skimage.transform import rotate, rescale
from skimage import data, measure
from skimage import filters
from skimage.filters import threshold_otsu
from skimage.feature  import canny

```

-----------------------------------------------------------------------------------------------------------------

```python
from skimage.feature import Cascade
trained_file = data.lbp_frontal_face_cascade_filename()

I = io.imread("face.jpg")

# Initialize the detector cascade
detector = Cascade(trained_file)

# Detect faces with min and max size of searching window
detected = detector.detect_multi_scale(img = I,
                                       scale_factor=1.2,
                                       step_ratio=1,
                                       min_size=(10, 10),
                                       max_size=(200, 200))


io.imshow(I)
plt.axis("off")
plt.title("original")
plt.show()
# Display the original image with detected faces highlighted
fig, ax = plt.subplots()
ax.imshow(I)
ax.axis('off')
ax.set_title("Original Image with Detected Faces")

# Draw bounding boxes around detected faces
for patch in detected:
    ax.add_patch(plt.Rectangle((patch['c'], patch['r']),
                               patch['width'], patch['height'],
                               linewidth=2, edgecolor='r', facecolor='none'))



```

```python
from skimage import io, transform
import matplotlib.pyplot as plt
import numpy as np
from skimage.color import rgb2gray
from skimage.transform import rotate, rescale
```

-----------------------------------------------------------------------------------------------------------------------------------

```python
# from skimage import morphology
#Import the data and the module needed for contouring detection.
# Obtain the horse image shown in the context area.
#Find the contours of the horse image using a constant level value of 0.8.
from skimage import data, measure

I = io.imread("horse.jpg")
horse_image = data.horse()

# Find the contours with a constant level value of 0.8
contours = measure.find_contours(horse_image, 0.8)

# See results
io.imshow(horse_image)
plt.axis("off")
plt.title("Original")
plt.show()

# Create a figure and axes for the contour plot
fig, ax = plt.subplots()
ax.imshow(horse_image, cmap=plt.cm.gray)  # Display the image in grayscale

# Plot the contours on top of the image
for contour in contours:
    ax.plot(contour[:, 1], contour[:, 0], linewidth=2)  # Plot x and y coordinates

ax.axis('image')  # Set aspect ratio to match image
ax.set_xticks([])  # Remove x-axis ticks
ax.set_yticks([])  # Remove y-axis ticks
plt.title("Contours")
plt.show()

```

```python
I = io.imread("dices.jpg")
from skimage.filters import threshold_otsu
from skimage import filters

from skimage import color
from skimage.color import rgb2gray


# Make the image grayscale
image_dice = color.rgb2gray(I)

# Obtain the optimal thresh value
thresh = filters.threshold_otsu(image_dice)

# Apply thresholding
binary = image_dice > thresh

# Find contours at a constant value of 0.8
contours = measure.find_contours(binary, 0.8)

io.imshow(I)
plt.axis("off")
plt.title("original")
plt.show()

# Create a figure and axes for the contour plot
fig, ax = plt.subplots()
ax.imshow(image_dice, cmap=plt.cm.gray)  # Display the image in grayscale

# Plot the contours on top of the image
for contour in contours:
    ax.plot(contour[:, 1], contour[:, 0], linewidth=2)  # Plot x and y coordinates

ax.axis('image')  # Set aspect ratio to match image
ax.set_xticks([])  # Remove x-axis ticks
ax.set_yticks([])  # Remove y-axis ticks
plt.title("Contours")
plt.show()
```
